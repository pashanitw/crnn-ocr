```python
_, ax = plt.subplots(2,4, figsize=(10, 5))
lookup = get_lookup()
for i, (images, labels) in enumerate(train_loader):
    for i in range(8):
        img = images[i].permute(1,2,0)
        img = (img).numpy()
        print("**************")
        print(labels[i])
        print(lookup.lookup_tokens(labels[i].tolist()))
        print("*********")
        #label = tf.strings.reduce_join(num_to_char(labels[i])).numpy().decode("utf-8")
        ax[i // 4, i % 4].imshow(img[:, :, 0], cmap="gray")
        ax[i // 4, i % 4].set_title(''.join(lookup.lookup_tokens(labels[i].tolist())))
        ax[i // 4, i % 4].axis("off")
    break
plt.show()
```
